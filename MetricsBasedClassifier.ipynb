```python
# K-Nearest Neighbors implementation
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Find the best k value
k_range = range(1, 21)
accuracies = []

for k in k_range:
    knn_temp = KNeighborsClassifier(n_neighbors=k)
    knn_temp.fit(X_train, y_train)
    y_pred_temp = knn_temp.predict(X_test)
    accuracies.append(accuracy_score(y_test, y_pred_temp))

# Plot k vs accuracy
plt.figure(figsize=(10, 6))
plt.plot(k_range, accuracies, marker='o')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.title('KNN: Finding Optimal k Value')
plt.grid(True)
plt.tight_layout()
plt.show()

best_k = k_range[np.argmax(accuracies)]
best_accuracy = max(accuracies)
print(f"Best k: {best_k} with accuracy: {best_accuracy:.4f}")

# Retrain with best k
KNN = KNeighborsClassifier(n_neighbors=best_k)
KNN.fit(X_train, y_train)
```

```python
# Make predictions
y_pred_knn = KNN.predict(X_test)
y_pred_proba_knn = KNN.predict_proba(X_test)[:, 1]
```

```python
# Evaluate the model
accuracy_knn = accuracy_score(y_test, y_pred_knn)
print(f"K-Nearest Neighbors Accuracy: {accuracy_knn:.4f}")
```

```python

print("\nClassification Report - KNN:")
print(classification_report(y_test, y_pred_knn, target_names=label_encoder.classes_))
```

```python
# Confusion Matrix
cm_knn = confusion_matrix(y_test, y_pred_knn)
print("\nConfusion Matrix - KNN:")
print(cm_knn)
```

```python
# Support Vector Machine implementation
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# Compare different SVM kernels
kernels = ['linear', 'rbf', 'poly']
svm_accuracies = []

for kernel in kernels:
    svm_temp = SVC(kernel=kernel, random_state=42, probability=True)
    svm_temp.fit(X_train, y_train)
    y_pred_temp = svm_temp.predict(X_test)
    accuracy_temp = accuracy_score(y_test, y_pred_temp)
    svm_accuracies.append(accuracy_temp)
    print(f"SVM with {kernel} kernel Accuracy: {accuracy_temp:.4f}")
```

```python
# Plot kernel comparison
plt.figure(figsize=(8, 5))
plt.bar(kernels, svm_accuracies, color=['blue', 'green', 'orange'], alpha=0.7)
plt.ylabel('Accuracy')
plt.title('SVM Performance with Different Kernels')
plt.ylim(0, 1)

# Add value labels
for i, acc in enumerate(svm_accuracies):
    plt.text(i, acc + 0.01, f'{acc:.4f}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Use the best kernel
best_kernel = kernels[np.argmax(svm_accuracies)]
print(f"\nBest SVM kernel: {best_kernel}")
```

```python
SVM = SVC(
    kernel=best_kernel,
    random_state=42,
    probability=True  # Enable probability estimates for consistency
)
```

```python
# Artificial Neural Network implementation using Pipeline
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Create ANN with specified architecture
ANN = MLPClassifier(
    hidden_layer_sizes=(24, 8),  # Two hidden layers: 24 and 8 neurons
    activation='logistic',  # Using relu for hidden layers (more common than sigmoid)
    solver='adam',
    random_state=42,
    max_iter=2000,
    learning_rate_init=0.01
)

# Create pipeline with scaler and ANN
# Note: We already scaled the data, but including in pipeline for completeness
ann_pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Scaler step
    ('ann', ANN)  # ANN model step
])

# Since we already scaled the data, we'll use the original unscaled data for pipeline
# Get the original unscaled data
X_train_original = scaler.inverse_transform(X_train) if hasattr(scaler, 'inverse_transform') else X_train
X_test_original = scaler.inverse_transform(X_test) if hasattr(scaler, 'inverse_transform') else X_test

# Train the ANN pipeline
ann_pipeline.fit(X_train_original, y_train)

# Make predictions
y_pred_ann = ann_pipeline.predict(X_test_original)
y_pred_proba_ann = ann_pipeline.predict_proba(X_test_original)[:, 1]

# Evaluate the model
accuracy_ann = accuracy_score(y_test, y_pred_ann)
print(f"Artificial Neural Network Accuracy: {accuracy_ann:.4f}")

print("\nClassification Report - ANN:")
print(classification_report(y_test, y_pred_ann, target_names=label_encoder.classes_))

# Confusion Matrix
cm_ann = confusion_matrix(y_test, y_pred_ann)
print("\nConfusion Matrix - ANN:")
print(cm_ann)

plt.figure(figsize=(6, 4))
sns.heatmap(cm_ann, annot=True, fmt='d', cmap='coolwarm',
            xticklabels=label_encoder.classes_,
            yticklabels=label_encoder.classes_)
plt.title('Confusion Matrix - Artificial Neural Network')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.show()

# Feature Importance for ANN (using permutation importance)
from sklearn.inspection import permutation_importance

# Calculate permutation importance
perm_importance_ann = permutation_importance(
    ann_pipeline, X_test_original, y_test,
    n_repeats=10,
    random_state=42
)

feature_importance_ann = pd.DataFrame({
    'feature': X_train.columns,
    'importance': perm_importance_ann.importances_mean,
    'std': perm_importance_ann.importances_std
}).sort_values('importance', ascending=False)

print("\nTop 3 Most Important Features for ANN (by permutation importance):")
print(feature_importance_ann.head(3))

# Plot Feature Importance
plt.figure(figsize=(10, 6))
top_features_ann = feature_importance_ann.head(3)
plt.barh(top_features_ann['feature'], top_features_ann['importance'],
         xerr=top_features_ann['std'], color='teal', alpha=0.7)
plt.xlabel('Permutation Importance Score')
plt.title('Top 3 Feature Importance - Artificial Neural Network')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# ANN Model Details
ann_model = ann_pipeline.named_steps['ann']
print("\nANN Model Details:")
print(f"Architecture: Input({ann_model.n_features_in_}) -> Hidden(24) -> Hidden(8) -> Output({ann_model.n_outputs_})")
print(f"Number of layers: {ann_model.n_layers_}")
print(f"Number of iterations: {ann_model.n_iter_}")
print(f"Loss: {ann_model.loss_:.4f}")
print(f"Activation function: {ann_model.activation}")
print(f"Output activation: softmax" if ann_model.out_activation_ == 'softmax' else f"Output activation: {ann_model.out_activation_}")

# Plot training loss curve
plt.figure(figsize=(8, 4))
plt.plot(ann_model.loss_curve_)
plt.title('ANN Training Loss Curve')
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Deep Neural Network implementation using MLPClassifier with 6 layers
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Create Deep Neural Network with specified 6-layer architecture
DNN = MLPClassifier(
    hidden_layer_sizes=(100, 70, 40, 24, 8),  # Five hidden layers: 100, 70, 40, 24, 8 neurons
    activation='relu',  # ReLU activation for hidden layers
    solver='adam',
    random_state=42,
    max_iter=2000,
    learning_rate_init=0.001,
    alpha=0.001,  # L2 regularization parameter
    early_stopping=True,
    n_iter_no_change=50,
    validation_fraction=0.1,
    beta_1=0.9,
    beta_2=0.999,
    epsilon=1e-8
)

# Create pipeline with scaler and DNN
dnn_pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Scaler step
    ('dnn', DNN)  # Deep Neural Network model step
])

# Get the original unscaled data
X_train_original = scaler.inverse_transform(X_train) if hasattr(scaler, 'inverse_transform') else X_train
X_test_original = scaler.inverse_transform(X_test) if hasattr(scaler, 'inverse_transform') else X_test

# Train the DNN pipeline
print("Training Deep Neural Network with 6 layers...")
dnn_pipeline.fit(X_train_original, y_train)

# Make predictions
y_pred_dnn = dnn_pipeline.predict(X_test_original)
y_pred_proba_dnn = dnn_pipeline.predict_proba(X_test_original)[:, 1]

# Evaluate the model
accuracy_dnn = accuracy_score(y_test, y_pred_dnn)
print(f"Deep Neural Network Accuracy: {accuracy_dnn:.4f}")

print("\nClassification Report - Deep Neural Network:")
print(classification_report(y_test, y_pred_dnn, target_names=label_encoder.classes_))

# Confusion Matrix
cm_dnn = confusion_matrix(y_test, y_pred_dnn)
print("\nConfusion Matrix - Deep Neural Network:")
print(cm_dnn)

plt.figure(figsize=(6, 4))
sns.heatmap(cm_dnn, annot=True, fmt='d', cmap='plasma',
            xticklabels=label_encoder.classes_,
            yticklabels=label_encoder.classes_)
plt.title('Confusion Matrix - Deep Neural Network (6 Layers)')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.tight_layout()
plt.show()

# Feature Importance for DNN (using permutation importance)
print("Calculating feature importance for Deep Neural Network...")
perm_importance_dnn = permutation_importance(
    dnn_pipeline, X_test_original, y_test,
    n_repeats=10,
    random_state=42
)

feature_importance_dnn = pd.DataFrame({
    'feature': X_train.columns,
    'importance': perm_importance_dnn.importances_mean,
    'std': perm_importance_dnn.importances_std
}).sort_values('importance', ascending=False)

print("\nTop 5 Most Important Features for Deep Neural Network:")
print(feature_importance_dnn.head(5))

# Plot Feature Importance
plt.figure(figsize=(12, 6))
top_features_dnn = feature_importance_dnn.head(5)
plt.barh(top_features_dnn['feature'], top_features_dnn['importance'],
         xerr=top_features_dnn['std'], color='darkviolet', alpha=0.7)
plt.xlabel('Permutation Importance Score')
plt.title('Top 5 Feature Importance - Deep Neural Network (6 Layers)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

# DNN Model Details
dnn_model = dnn_pipeline.named_steps['dnn']
print("\nDeep Neural Network Model Details:")
print(f"Architecture: Input({dnn_model.n_features_in_}) -> Hidden(100) -> Hidden(70) -> Hidden(40) -> Hidden(24) -> Hidden(8) -> Output({dnn_model.n_outputs_})")
print(f"Total layers: {dnn_model.n_layers_}")
print(f"Total parameters: ~{sum(layer.size for layer in dnn_model.coefs_):,}")
print(f"Number of iterations: {dnn_model.n_iter_}")
print(f"Final loss: {dnn_model.loss_:.4f}")
print(f"Hidden layer activation: {dnn_model.activation}")
print(f"Output activation: {'softmax' if dnn_model.out_activation_ == 'softmax' else dnn_model.out_activation_}")

```
